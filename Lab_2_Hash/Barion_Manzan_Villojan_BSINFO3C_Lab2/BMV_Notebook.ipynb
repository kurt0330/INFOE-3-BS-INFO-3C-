```python
import hashlib

filename = 'sampleDB.csv' 

def get_file_hash(path):
    
    sha256_hash = hashlib.sha256()
    with open(path, "rb") as f:
        
        for byte_block in iter(lambda: f.read(4096), b""):
            sha256_hash.update(byte_block)
    return sha256_hash.hexdigest()

# Generate and print the result
try:
    
    print(f"File: {filename}")
    print(f"SHA-256 Hash: {get_file_hash(filename)}") # Copy the hash and paste it into 'EXPECTED_HASH' variable.
    print("-" * 30)
    
except FileNotFoundError:
    print(f"Error: '{filename}' not found.")
```

```python
import hashlib
import os

# STEP 1: The 'Protection Profile' 
# Paste the long string from Cell 1 inside the quotes below.
EXPECTED_HASH = "656509a3199176a4559644316ece7395c41d3237e97a6019b77c5ddc209d5d6908" 
filename = 'sampleDB.csv'  # Ensure this matches your file name

def run_integrity_audit():
    # A. THE AVAILABILITY CHECK
    # Check if the file exists before trying to hash it
    if not os.path.exists(filename):
        print(f"ALERT: File '{filename}' not found! Availability has been compromised.")
        return

    # B. THE FINGERPRINTING
    sha256_hash = hashlib.sha256()
    
    try:
        # Open the file in 'rb' (read binary) mode
        with open(filename, "rb") as f:
            # Loop to read the file in small blocks (4096 bytes)
            for byte_block in iter(lambda: f.read(4096), b""):
                sha256_hash.update(byte_block)
        
        # Convert the result to a '.hexdigest()'
        current_hash = sha256_hash.hexdigest()
        
        print(f"Current Hash:  {current_hash}")
        print(f"Expected Hash: {EXPECTED_HASH}")
        print("-" * 30)

        # C. THE TECHNICAL CONTROL
        # Compare the new hash to the EXPECTED_HASH
        if current_hash == EXPECTED_HASH:
            # What message should the Subject (the script) print if they match?
            # Message if they match (Integrity Intact)
            print("SUCCESS: Integrity Verified. The file has not been modified.")
        else:
            # What alert should it trigger if they don't?
            # Alert if they don't match (Integrity Lost)
            print("!!! ALERT !!!: INTEGRITY LOST!")
            print("The file has been modified or corrupted by a Threat Agent.")
            
    except Exception as e:
        print(f"An error occurred during the audit: {e}")

# Run the function
run_integrity_audit()
```
